# Drop Uncessary Columns

```python
df.drop(['ranking of movie', 'certificate', 'ACTOR 2', 'ACTOR 3', 'ACTOR 4', 'DETAIL ABOUT MOVIE'], axis=1 , inplace=True)
```

```python
df.head()
```

```python
df.columns
```

# Rename columns in new Form
- remove spaces, uppercase and white spaces

```python
df.columns
```

```python
#make cols heading lowercase
df.columns = df.columns.str.lower()
```

```python
df.columns
```

```python
#remove white spaces
df.columns = df.columns.str.strip()
```

```python
df.columns
```

```python
#replace in between spaces by underscore
df.columns = df.columns.str.replace(' ', '_')
```

```python
df.columns
```
