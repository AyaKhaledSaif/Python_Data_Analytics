# Supermarket Sales Exercise - Solutions

Welcome to a quick exercise for you to practice your visualization skills! We will be using the [supermarket-sales Dataset](https://www.kaggle.com/aungpyaeap/supermarket-sales) from Kaggle! Just follow along and complete the tasks outlined in bold below. The tasks will get harder and harder as you go along.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(rc={'figure.figsize': [9, 9]}, font_scale=1.2)
```

```python
df = pd.read_csv('supermarket_sales - Sheet1.csv')
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
sns.distplot(df['Rating'], kde=False, bins=30, color='m')
```

```python
sns.kdeplot(df['Rating'], shade=True, color='m')
```

```python
sns.distplot(df['Tax 5%'], kde=False, bins=30, color='m')
```

```python
sns.kdeplot(df['Tax 5%'], shade=True, color='m')
```

```python
sns.countplot(x='Gender', data=df)
```

```python
plt.figure(figsize=(6,4))
sns.countplot(y='Product line', data=df, palette='rocket')
```

```python
sns.countplot(y='Product line', data=df, hue='Gender',palette='viridis')
```

```python
sns.countplot(x='Quantity', data=df, hue='Gender',palette='BuPu')
```

```python
sns.countplot(x='Branch', data=df, hue='Customer type')
```

```python
sns.countplot(x='Payment',hue='Gender', data=df, palette='Set1')
```

```python
sns.boxplot(x='Gender', y='Total', data=df)
```

```python
sns.boxplot(y='Product line', x='Total', data=df)
```

```python
sns.boxplot(x='Branch', y='Total', data=df)
```

```python
sns.violinplot(x='City', y='gross income', data=df, hue='Gender', split=True)
```

```python
sns.stripplot(x='Payment', y='Total', data=df, hue='Branch', dodge=True)
```

```python
sns.swarmplot(x='Payment', y='Total', data=df, hue='Branch', dodge=True)
```

```python
sns.violinplot(x='Payment', y='Tax 5%', data=df, hue='Gender', split=True)
```

```python
sns.barplot(y='Product line', x='Quantity', data=df, estimator=np.sum, hue='Gender')
```
