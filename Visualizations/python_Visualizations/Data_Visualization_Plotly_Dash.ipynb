# What is Data Visualization

Data visualization is the graphical representation of information and data. By using visual elements like charts, graphs, and maps, data visualization tools provide an accessible way to see and understand trends, outliers, and patterns in data.
- Plotly stewards Python's leading data viz and UI libraries. 
- With Dash Open Source, you can create data apps on your laptop in pure Python, no JavaScript required.

Common Types of Data Plots:
- Bar chart
- Line plot
- Scatter plot
- Histogram
- Box plot
- Pie chart

Advanced Types of Data Plots:
- Grouped bar chart
- Stacked area chart
- Stacked bar chart
- Multiple box plot
- Violin plot
- Heatmap

8. Stacked Area Plot on flights dataset

```python
# Number of passengers in 1949 and 1950 by month
flights.head()
```

```python
flights.info()
```

- In Python, the isin function is commonly used with pandas, a powerful data manipulation library. 
- The isin function is used to filter data frames based on whether values are present in a specified list or another data structure.

```python
flights[flights['year'].isin([1949,1950])].groupby(['year', 'month'])['passengers'].sum().reset_index()
```

```python
# Filter data for the years 1949 and 1950
flights_4950 = flights[flights['year'].isin([1949, 1950])]

# Group by year and month, and calculate the sum of passengers
passengers_by_month = flights_4950.groupby(['year', 'month'])['passengers'].sum().reset_index()

# Print the result
print(passengers_by_month)
```

- In Python, a pivot table is a data processing technique used to summarize, aggregate, and reorganize data in a tabular form. 
- The pandas library provides a convenient pivot_table function that allows you to create pivot tables from DataFrames. 
- Pivot tables are commonly used in data analysis and reporting to analyze and present data in a more structured format.

```python
import pandas as pd

# Sample DataFrame
data = {
    'Date': ['2022-01-01', '2022-01-01', '2022-01-02', '2022-01-02'],
    'Category': ['A', 'B', 'A', 'B'],
    'Value': [10, 15, 20, 25]
}

df = pd.DataFrame(data)

# Pivot table example
pivot_table = df.pivot_table(index='Date', columns='Category', values='Value', aggfunc='sum')

print("Original DataFrame:")
print(df)

print("\nPivot Table:")
print(pivot_table)
```

```python
# Pivot the data to create a DataFrame suitable for a stacked area chart
pivot_df = flights_4950.pivot(index='month', columns='year', values='passengers').reset_index()
pivot_df
```

```python
plt.stackplot(pivot_df['month'], pivot_df[1949], pivot_df[1950], labels=['1949', '1950'], alpha=0.7)
plt.title('Stacked Area Chart of Passengers in 1949 and 1950 by Month')
plt.xlabel('Month')
plt.ylabel('Number of Passengers')
plt.legend(loc='upper left')
plt.show()
```

```python
#Draw using line graph

# Filter data for the years 1949 and 1950
flights_4950 = flights[flights['year'].isin([1949, 1950])]

# Plot the number of passengers by month
plt.figure(figsize=(12, 6))
sns.lineplot(x='month', y='passengers', hue='year', data=flights_4950, marker='o')
plt.title('Number of Passengers in 1949 and 1950 by Month')
plt.xlabel('Month')
plt.ylabel('Number of Passengers')
plt.legend(title='Year')
plt.show()
```

9. Plot Stacked Bar chart on flight dataset

```python
# Create a stacked bar chart
plt.figure(figsize=(12, 6))
plt.bar(pivot_df['month'], pivot_df[1949], label='1949', alpha=0.7)
plt.bar(pivot_df['month'], pivot_df[1950], bottom=pivot_df[1949], label='1950', alpha=0.7)

# Adding labels and title
plt.title('Stacked Bar Chart of Passengers in 1949 and 1950 by Month')
plt.xlabel('Month')
plt.ylabel('Number of Passengers')
plt.legend(loc='upper left')
plt.show()
```

12. Heatmap plot on car_crashes dataset

```python
# Data preparation
from sklearn import preprocessing
car_crashes_cleaned = car_crashes.drop(labels='abbrev', axis=1).iloc[0:10]
min_max_scaler = preprocessing.MinMaxScaler()
car_crashes_normalized = pd.DataFrame(min_max_scaler.fit_transform(car_crashes_cleaned.values), columns=car_crashes_cleaned.columns)

# Creating a heatmap
sns.heatmap(car_crashes_normalized, annot=True)
plt.title('Car crash heatmap for the first 10 car crashes')
plt.show()
```
